Taejeong Bae created filtered callsets for some of our CommonMind samples.  Here we look at the human clinical metadata for those samples.

```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import synapseclient
import re
```

Loging to Synapse and read [CMC_Human_clinical_metadata.csv](https://www.synapse.org/#!Synapse:syn2279441) (syn2279441) into a pandas DataFrame.

```python
syn = synapseclient.login()
e = syn.get('syn2279441', downloadLocation='/home/attila/projects/bsm/results/2020-04-08-cmc-clinical-metadata-tj/')
cmc = pd.read_csv(e.path, index_col='Individual ID')
cmc
```

List Taejeong's VCF files in Synapse folder [WGS.SCZ.Chess](https://www.synapse.org/#!Synapse:syn21897893) (syn21897893).

```python
vcfs_tj = [y['name'] for y in syn.getChildren('syn21897893')]
synids_tj = [y['id'] for y in syn.getChildren('syn21897893')]
vcfs_tj[:3]
```

Derive Individual IDs from the VCF file names and extract the corresponding samples/individuals from CMC_Human_clinical_metadata.csv

```python
ids_tj = [re.sub('^((MSSM|PITT)_[0-9]+).*$', 'CMC_\\1', y) for y in vcfs_tj]
tj = cmc.loc[ids_tj, ['Dx', 'ageOfDeath', 'Reported Gender', 'Sex']]
tj['VCF'] = vcfs_tj
tj['Synapse ID'] = synids_tj
tj.to_csv('/home/attila/projects/bsm/results/2020-04-08-cmc-clinical-metadata-tj/CMC_Human_clinical_metadata-TJ.csv')
tj
```
