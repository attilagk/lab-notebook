```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot()
```

Einstein proposed that

$$
\begin{eqnarray}
E &=& mc^2 \\
c &\approx& 3\times 10^5 \frac{\mathrm{km}}{\mathrm{s}}
\end{eqnarray}
$$

Significant points of this formula
* energy $E$ and mass $m$ are inseparable from each other
* bodies impact each other depending on their relative speed (relativity)
